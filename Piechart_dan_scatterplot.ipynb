# Explaratory Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('trst1.csv',
                 skipfooter=2)
```

```python
df['Total'] = df.sum(axis=1)

df.head(10)
```

```python
df = df.groupby('Name of Countries', axis=0).sum()
df
```

# PIE CHART

```python
df5 = df.head()

df5
```

```python
colors_list = ['red', 'yellow', 'lightcoral', 'green', 'blue']
explode_list = [0.1, 0, 0.1, 0, 0]
df5['Total'].plot(kind='pie',
                            figsize=(15, 6),
                            autopct='%1.1f%%', 
                            startangle=90,    
                            shadow=True,       
                            labels=None,         # turn off labels on pie chart
                            pctdistance=1.12,    # the ratio between the center of each pie slice and the start of the text generated by autopct 
                            colors=colors_list,
                            explode=explode_list# add custom colors)
)
                            

# scale the title up by 12% to match pctdistance
plt.title("Piechart untuk Tingkat Kedatangan Turis di India", y=1.12) 

plt.axis('equal') 

# add legend
plt.legend(labels=df5.index, loc='upper left') 

plt.show()
```

# BARPLOT

```python
years = list(map(str, range(2001,2015)))
```

```python
df_countries = df.loc[['USA'],years].transpose()
df_countries
```

```python
 df_countries.plot(kind='box', figsize=(8, 6))

plt.title("Boxplot untuk Tingkat Kedatangan Turis dari USA ke India")
plt.ylabel("Tingkat Kedatangan")

plt.show()
```

```python
df_countries = df.loc[['USA', 'CANADA'],years].transpose()
df_countries
```

```python
df_countries.plot(kind='box', figsize=(8, 6))

plt.title("Boxplot untuk Tingkat Kedatangan Turis dari USA dan Canada ke India")
plt.ylabel("Tingkat Kedatangan")

plt.show()
```

# SCATTER PLOT

```python
df_countries.describe()
```

```python
# we can use the sum() method to get the total population per year
df_tot = pd.DataFrame(df[years].sum(axis=0))

# change the years to type int (useful for regression later on)
df_tot.index = map(int, df_tot.index)

# reset the index to put in back in as a column in the df_tot dataframe
df_tot.reset_index(inplace = True)

# rename columns
df_tot.columns = ['years', 'total']

# view the final dataframe
df_tot.head()
```

```python
df_tot.plot(kind='scatter', x='years', y='total', figsize=(10, 6), color='darkblue')

plt.title('Boxplot untuk Tingkat Kedatangan Turis dari USA dan Canada ke India')
plt.xlabel('Years')
plt.ylabel('Tingkat Kedatangan')
plt.show()
```

```python
x = df_tot['years']      # year on x-axis
y = df_tot['total']     # total on y-axis
fit = np.polyfit(x, y, deg=1)

fit
```

```python
df_tot.plot(kind='scatter', x='years', y='total', figsize=(10, 6), color='darkblue')

plt.title('Boxplot untuk Tingkat Kedatangan Turis dari USA dan Canada ke India')
plt.xlabel('Years')
plt.ylabel('Tingkat Kedatangan')

# plot line of best fit
plt.plot(x, fit[0] * x + fit[1], color='red') # recall that x is the Years
plt.annotate('y={0:.0f} x + {1:.0f}'.format(fit[0], fit[1]), xy=(2000, 935))

plt.show()

# print out the line of best fit
'No. Consumption = {0:.0f} * Year + {1:.0f}'.format(fit[0], fit[1]) 
```
