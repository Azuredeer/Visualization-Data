# WAFFLE CHART

```python
df_cn3 = df.head(3)

df_cn3
```

```python
# compute the proportion of each category with respect to the total
total_values = df_cn3['Total'].sum()
category_proportions = df_cn3['Total'] / total_values

# print out proportions
pd.DataFrame({"Category Proportion": category_proportions})
```

```python
width = 40 # width of chart
height = 10 # height of chart

total_num_tiles = width * height # total number of tiles

print(f'Total number of tiles is {total_num_tiles}.')
```

```python
# compute the number of tiles for each category
tiles_per_category = (category_proportions * total_num_tiles).round().astype(int)

# print out number of tiles per category
pd.DataFrame({"Number of tiles": tiles_per_category})
```

```python
# initialize the waffle chart as an empty matrix
waffle_chart = np.zeros((height, width), dtype = np.uint)

# define indices to loop through waffle chart
category_index = 0
tile_index = 0

# populate the waffle chart
for col in range(width):
    for row in range(height):
        tile_index += 1

        # if the number of tiles populated for the current category is equal to its corresponding allocated tiles...
        if tile_index > sum(tiles_per_category[0:category_index]):
            # ...proceed to the next category
            category_index += 1       
            
        # set the class value to an integer, which increases with class
        waffle_chart[row, col] = category_index
        
print ('Waffle chart populated!')
```

```python
waffle_chart
```

```python
# instantiate a new figure object
fig = plt.figure()

# use matshow to display the waffle chart
colormap = plt.cm.coolwarm
plt.matshow(waffle_chart, cmap=colormap)
plt.colorbar()
plt.show()
```

```python
# instantiate a new figure object
fig = plt.figure()

# use matshow to display the waffle chart
colormap = plt.cm.coolwarm
plt.matshow(waffle_chart, cmap=colormap)
plt.colorbar()

# get the axis
ax = plt.gca()

# set minor ticks
ax.set_xticks(np.arange(-.5, (width), 1), minor=True)
ax.set_yticks(np.arange(-.5, (height), 1), minor=True)
    
# add gridlines based on minor ticks
ax.grid(which='minor', color='w', linestyle='-', linewidth=2)

plt.xticks([])
plt.yticks([])
plt.show()
```

# WORD CLOUDS

```python
# install wordcloud
!pip3 install wordcloud==1.8.1

# import package and its set of stopwords
from wordcloud import WordCloud, STOPWORDS

print ('Wordcloud is installed and imported!')
```

```python
import urllib

# open the file and read it into a variable frozen
frozen = urllib.request.urlopen('https://raw.githubusercontent.com/Azuredeer/Visualization-Data/main/Tugas%204/Frozen.txt').read().decode("utf-8")
```

```python
stopwords = set(STOPWORDS)
```

```python
# instantiate a word cloud object
frozen_wc = WordCloud(
    background_color='white',
    max_words=2000, #took 2000 words on first teks
    stopwords=stopwords
)

# generate the word cloud
frozen_wc.generate(frozen)
```

```python
fig = plt.figure(figsize=(14, 18))

# display the cloud
plt.imshow(frozen_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
stopwords.add('said') # add the words said to stopwords

# re-generate the word cloud
frozen_wc.generate(frozen)

# display the cloud
fig = plt.figure(figsize=(14, 18))

plt.imshow(frozen_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# save mask to alice_mask
elsa_mask = np.array(Image.open(urllib.request.urlopen('https://raw.githubusercontent.com/Azuredeer/Visualization-Data/main/Tugas%204/Frozen.jpg')))
```

```python
fig = plt.figure(figsize=(14, 18))

plt.imshow(elsa_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# instantiate a word cloud object
frozen_wc = WordCloud(background_color='white', max_words=2000, mask=elsa_mask, stopwords=stopwords)

# generate the word cloud
frozen_wc.generate(frozen)

# display the word cloud
fig = plt.figure(figsize=(14, 18))

plt.imshow(frozen_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
