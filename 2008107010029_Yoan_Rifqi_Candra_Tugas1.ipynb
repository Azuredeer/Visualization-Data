```python
# in pandas axis=0 represents rows (default) and axis=1 represents columns.
df.drop(['Currency'], axis=1, inplace=True)
```

```python
df.head(3)
```

```python
df.rename(columns={'Date':'Tanggal', 'Open':'Buka', 'High':'Tertinggi', 'Low':'Terendah', 'Close':'Tutup'}, inplace=True)
df.columns
```

```python
df['Total'] = df.sum(axis=1, numeric_only=True)
```

```python
df['Tanggal'] = pd.to_datetime(df['Tanggal'])
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
mona = pd.DataFrame(df, columns=['Tanggal','Tertinggi'])
sns.lineplot(x='Tanggal', y='Tertinggi', data = mona)
```

```python
df.plot(x='Tanggal', y=['Tertinggi', 'Terendah', 'Buka', 'Tutup'])
```

```python

df['Tanggal']=pd.to_datetime(df["Tanggal"])
df.set_index('Tanggal',inplace=True)
df.index = pd.DatetimeIndex(df.index)

```

```python
df['2021':]['Buka'].plot(figsize=(12,8))
plt.legend(bbox_to_anchor=(1.25, 0.5))
plt.title("Data Harga Buka")
sns.despine()
```
